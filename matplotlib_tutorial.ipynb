```python
from numpy import pi, exp, linspace, abs
from scipy.special import jv # Bessel function

from matplotlib import rc # for latex
rc('text', usetex=True)
rc('text.latex', preamble=r'\usepackage{siunitx}') # need that for non-italic µ


############ Constants ###########################
H_TARGET = 75 # target harmonic of the seed laser1
N1 = 1 # harmonic of the first seed laser
N2 = 2 # harmonic of the second seed laser
LAMBDA_0 = 300e-9 # common wavelength, m
K_0 = 2*pi/LAMBDA_0 # common wavenumber, m^{-1}
SIGMA_ETA = 0.150/1350 # initial energy spread
ETA_M1 = 0.452/1350 # relative energy spread in the first modulator
ETA_M2 = 0.750/1350 # relative energy spread in the second modulator
R2_SIGN = 1 # sign of dispersion in the second chicane

R1_range = linspace(6.5e-3, 7.5e-3, 100) # range for dispersion of the first chicane
R2_range = linspace(75e-6, 100e-6, 100) # range for dispersion of the second chicane

def get_bunching(R1, R2, h=H_TARGET, n1=N1, n2=N2, k_0=K_0, \
                 eta_m1=ETA_M1, r2_sign=R2_SIGN, eta_m2=ETA_M2, sigma_eta=SIGMA_ETA):
    res = 0 # to store the result
    for n in range(-25, 25): # reasonable summation range, make smaller if takes too long
        C1 = k_0*(h*R2-n1*n*R1) # temp param
        C2 = h*k_0*R2 # temp param
        envelope = exp(-C1**2*sigma_eta**2/2) # temp param
        m = round((h + n1*n)/n2) # temp param
        res += (-1)**(m)*jv(m, C2*eta_m2)*jv(n, C1*eta_m1)*envelope # the main formula
    return abs(res) # takes amplitude of (potentially) complex number


matrix = [] # to store the data for plt.imshow()
for r1 in R1_range: # will be your vertical axis
    line = [] # temp storage for single line in the matrix
    for r2 in R2_range: # will be your horizontal axis
        line.append(get_bunching(r1, r2)) # writes the line
    matrix.append(line) # adds the line to the matrix
########################################################

fig, ax = plt.subplots(figsize=(7, 7), dpi = 80)    
pos = ax.imshow(matrix, cmap='hot', interpolation='bilinear') # interpolation option - 'bilinear'

# Ticks are tricky, because once you calculated the matrix, python doesn't remember,
# to which pair of values R1 and R2 each element of the matrix corresponds
# One way to go is the following
yticks = linspace(R1_range[0], R1_range[-1], 10) # choose ten points inside R1_range
yticks = [round(i*1e3, 2) for i in yticks] # convert from m to mm
ax.set_yticks(linspace(0, len(R1_range), 10))# put 10 ticks on y axis and put your custom ticks on those ticks
ax.set_yticklabels(yticks)
# The same for x-axis
xticks = linspace(R2_range[0], R2_range[-1], 10) # choose ten points inside R2_range
xticks = [round(i*1e6) for i in xticks] # convert from m to µm
ax.set_xticks(ticks=linspace(0, len(R2_range), 10))
ax.set_xticklabels(xticks)

ax.set_ylabel(r'$R_{56}^{(1)},$ mm', fontsize = 18)
ax.set_xlabel(r'$R_{56}^{(2)},$ \SI{}{\micro m}', fontsize = 18) # to write microns properly


# create an axes on the right side of ax. The width of cax will be 5%
# of ax and the padding between cax and ax will be fixed at 0.05 inch.
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.05)

plt.colorbar(pos, cax=cax)
#fig.colorbar(pos, ax=ax, shrink=0.88)

plt.show()
```

### 3d plot
For the surface topology visualization, fields visualization etc. 

```python
plt.clf()
from mpl_toolkits.mplot3d import Axes3D
import cmocean #for heatmap

X4,Y4= np.meshgrid(np.linspace(-1, 1, 25), np.linspace(-1, 1, 25)) #Return evenly spaced numbers over a specified interval.
Z4 = np.sin(5*np.sqrt(X4**2+Y4**2))

# creating a 3D-plot

fig, ax = plt.subplots(subplot_kw={"projection": "3d"}, figsize=(12,8), dpi = 100)
im = ax.plot_surface(X4, Y4, Z4, rstride=1, cstride=1, alpha = 0.95,\
                      cmap=cmocean.cm.thermal, linewidth=.25) #rstride - Array row stride (step size), cstride - Array column stride (step size)

# Plotting contour lines
cset = ax.contour(X4, Y4, Z4, zdir='z', offset = -1.05, linewidths = 0.3, cmap = "hot")
cset = ax.contour(X4, Y4, Z4, zdir='y', offset = -1.05, linewidths = 0.3, cmap = "hot")
cset = ax.contour(X4, Y4, Z4, zdir='x', offset = -1.05, linewidths = 0.3, cmap = "hot")

ax.set_xlim3d(np.min(X4),np.max(X4));
ax.set_ylim3d(np.min(Y4),np.max(Y4));
ax.set_zlim3d(np.min(Z4),np.max(Z4));

ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])

#fig.colorbar(im, shrink=0.5, aspect=5)

ax.view_init(32, 60) #'elev' stores the elevation angle in the z plane. 'azim' stores the azimuth angle in the x,y plane
plt.savefig('threed.png', bbox_inches = 'tight', transparent = True)
```
